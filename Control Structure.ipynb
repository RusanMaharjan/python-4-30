## IF-Else Practice Codes

```python
# Write a program to show temperature is hot when temperature celcius is more than 25 degree.
# Flow of program is waterfall method.
temperature = 23

if temperature >= 25: # 23 >= 25  # False
    print('Its hot today!!')
else:
    print('Its cold today!!')
```

```python
temperature = float(input('Enter temperature value: '))

if temperature >= 30: # 32 >= 30  # True
    print('Its hot today!!')
elif temperature >= 20 and temperature < 30: # 18 >= 20 -> False and 18 < 30 -> True -> False
    print('Its warm today!!')
else:
    print('Its cold today!!')
```

```python
# Student Grade Calculation
# 5 subject marks input
# total marks
# average marks %

sub1 = float(input('Enter subject 1 marks: '))
sub2 = float(input('Enter subject 2 marks: '))
sub3 = float(input('Enter subject 3 marks: '))
sub4 = float(input('Enter subject 4 marks: '))
sub5 = float(input('Enter subject 5 marks: '))

total_marks = sub1 + sub2 + sub3 + sub4 + sub5
print(f'\nTotal Marks: {total_marks}')

marks = total_marks / 5
print(f'Average Marks: {marks}')

if marks >= 80 and marks <= 100:
    print('Grade: A+')
elif marks >= 70 and marks < 80:
    print('Grade: A')
elif marks >= 60 and marks < 70:
    print('Grade: B')
elif marks >= 50 and marks < 60:
    print('Grade: C')
elif marks >= 40 and marks < 50:
    print('Grade: D')
elif marks >= 0 and marks < 40:
    print('Fail')
else:
    print('Invalid Input..')
```

### Tax Calculator

```python
annual_income = int(input('Enter annual income: '))
tax_amount = 0

# 0 - 250000-
if annual_income >=0 and annual_income < 250000:
    print('No Tax')
# 250000 - 500000
elif annual_income >= 250000 and annual_income <= 500000:
    print(f'Total Annual Income: Rs.{annual_income}')
    tax_amount = annual_income * 0.05 # (annual_income * 5) / 100
    annual_income -= tax_amount # annual_income = annual_income - tax_amount
    print(f'Tax to be deducted (5%): Rs.{tax_amount}')
    print(f'Remaining Annual Income after Tax: {annual_income}')
# 500000 - 1000000
elif annual_income > 500000 and annual_income <= 1000000:
    print(f'Total Annual Income: Rs.{annual_income}')
    tax_amount = annual_income * 0.10
    annual_income -= tax_amount # annual_income = annual_income - tax_amount
    print(f'Tax to be deducted (10%): Rs.{tax_amount}')
    print(f'Remaining Annual Income after Tax: {annual_income}')
# > 1000000
elif annual_income > 1000000:
    print(f'Total Annual Income: Rs.{annual_income}')
    tax_amount = annual_income * 0.20
    annual_income -= tax_amount # annual_income = annual_income - tax_amount
    print(f'Tax to be deducted (20%): Rs.{tax_amount}')
    print(f'Remaining Annual Income after Tax: {annual_income}')
else:
    print('Invalid Annual Income!!')
```

```python
pin = 1234
balance = 50000

user_pin = int(input('Enter your pin: '))
amount = int(input('Enter your withdraw balance: '))

if (user_pin == pin) and (amount < balance) and (amount % 1000 == 0):
    balance -= amount
    print(f'Rs.{amount} withdrawn from A/C 2134567890.')
    print(f'Rs.{balance} remaining in your account.')
elif (user_pin != pin) and (amount < balance) and (amount % 1000 == 0):
    print(f'Invalid Pin. Pin Number doesnot match.')
elif (user_pin == pin) and (amount >= balance) and (amount % 1000 == 0):
    print(f'Invalid balance amount. Withdrawn amount is more than balance amount.')
elif (user_pin == pin) and (amount < balance) and (amount % 1000 != 0):
    print(f'Amount value must be multiple of 1000s.')
else:
    print(f'Invalid Value!!')
```

## Nested If

```python
# 1. student attedance >= 75
# 2. student internal_marks >= 40
```

```python
attendance = int(input('Enter attendance: '))
internal_marks = float(input('Enter internal marks: '))

if attendance >= 75 and internal_marks >= 40:
    print('You are eligible for final examination.')
elif attendance < 75 and internal_marks >= 40:
    print('You not are eligible for final examination. Attendance requirement doesnot meet.')
elif attendance >= 75 and internal_marks < 40:
    print('You not are eligible for final examination. Internal Marks requirement doesnot meet.')
else:
    print('Invalid value!!')
```

```python
attendance = int(input('Enter attendance: '))

if attendance >= 75: # False
    internal_marks = float(input('Enter internal marks: '))
    if internal_marks >= 40: # True
        print('You are eligible for final examination.')
    else:
        print('You not are eligible for final examination. Internal Marks requirement doesnot meet.')
else:
    print('You not are eligible for final examination. Attendance requirement doesnot meet.')
```

```python
pin = 1234
balance = 50000

user_pin = int(input('Enter your pin number: '))

if user_pin == pin:
    amount = int(input('Enter your withdraw amount: '))
    if amount < balance:
        if amount % 1000 == 0:
            balance -= amount
            print(f'Rs.{amount} withdrawn from A/C 2134567890.')
            print(f'Rs.{balance} remaining in your account.')
        else:
            print(f'Withdraw amount must be multiple of 1000s.')
    else:
        print('Withdraw balance must be less than balance amount.')
else:
    print('Invalid Pin Number. Please enter valid pin number.')
```
