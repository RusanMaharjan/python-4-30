```python
'''
When to use Return type function?
-> If we need to call our function from one file to another then use reutrn type function.
-> we cannot use print function to display output. Just need to call function.
-> Only one return statement can be defined in one single function.

When to use non return type function?
-> If we need to call our function in same file where we defined it then use non return type function.
-> we can use print function to display output.
-> Multiple print statement can be defined in one single function.
'''
```

## Default Argument Function

```python
def productDetails(product_name='Unknown', product_desc='Unknown Description', 
                   product_price=0, product_quantity=0):
    print(f'Product Name: {product_name}')
    print(f'Product Description: {product_desc}')
    print(f'Product Price: {product_price}')
    print(f'Product Quantity: {product_quantity}')
    print(f'Total Price: {product_price * product_quantity}')
```

```python
productDetails(100000)
```

## Keyword Argument Function

```python
def productDetails(product_name='Unknown', product_desc='Unknown Description', 
                   product_price=0, product_quantity=0):
    print(f'Product Name: {product_name}')
    print(f'Product Description: {product_desc}')
    print(f'Product Price: {product_price}')
    print(f'Product Quantity: {product_quantity}')
    print(f'Total Price: {product_price * product_quantity}')
```

```python
productDetails(product_quantity=10, product_name='Mobile', 
               product_desc = 'Black Mobile', product_price=2000)
```

## All Argument Function

```python
# *args
def personalDetails(name, *args):
    for data in args:
        print(data)
    print(name)
```

```python
personalDetails('Ram', 30, 'Bachelors', 'Lalitpur', 50000)
```

## All Keyword Argument Function

```python
# **kwargs
def kwPersonalDetails(**kwargs):
    print(kwargs)
    for key, val in kwargs.items():
        print(f'{key}: {val}')
```

```python
kwPersonalDetails(name='Ram', age=30, qualification= 'Bachelors', address='Lalitpur', salary=50000)
```

## Nested Function

```python
# Multiple independent function / single independent function called by one single function is nested function.
def add(a, b):
    return f'Addition of {a} and {b} is: {a + b}.'

def sub(a, b):
    return f'Subtraction of {a} and {b} is: {a - b}'

def mul(a, b):
    return f'Multiplication of {a} and {b} is: {a * b}'

def div(a, b):
    return f'Division of {a} and {b} is: {a / b}.'
```

```python
def calculator(a, b):
    return add(a, b), sub(a, b), mul(a, b), div(a, b)
```

```python
calculator(1, 2)
```

```python
def calc(choice):
    num1 = float(input('Enter a number: '))
    num2 = float(input('Enter a number: '))
    if choice == 1:
        selected_choice = add(num1, num2)
    elif choice == 2:
        selected_choice = sub(num1, num2)
    elif choice == 3:
        selected_choice = mul(num1, num2)
    elif choice == 4:
        selected_choice = div(num1, num2)
    else:
        print('invalid choice')

    return selected_choice
```

```python
calc(2)
```

```python
def odd_even(num):
    if num % 2 == 0:
        print('Even')
    else:
        print('Odd')
odd_even(4)
```

```python
o_e(2)
```

```python
prices = [100, 250, 75, 400, 120, 310]
cheap_items = []
expensive_items = []
```

```python
def get_exp(lst):
    for price in lst:
        if price >= 200:
            expensive_items.append(price)


def get_che(lst):
    for price in lst:
        if price < 200:
            cheap_items.append(price)

get_exp(prices)
get_che(prices)
```

```python
get_expensive = lambda lst : [price for price in lst if price >= 200]
get_cheap = lambda lst : [price for price in lst if price < 200]
```

```python
get_expensive(prices)
```

```python
get_cheap(prices)
```
