# Functions

```python
'''
- Function is like a component.
- Write code inside function once and run it to multiple set of code at once.

syntax
-------
def  -> define

def function_name():
    logic..
    print(...)
    return...
'''
```

```python
'''
Types of function
-----------------------
1. Argumented / Parameterized Functions
    a. Default Argument Function
    b. Keyword Argument Function
    c. All Argument Function (*args)
    d. All Keyword Argument Function (**kwargs)
    
2. Non Arugmented / Non Parameterized Functions
3. Return Type and Non Return Type Functions
4. Lambda Function
'''
```

```python
# Non-Argumented and non return type function
def greet():
    print('Hello World!!')
    print('ABCDEFGHIJKL')
    print('ABCDEFGHIJKL')
    print('ABCDEFGHIJKL')
    print('ABCDEFGHIJKL')
    print('ABCDEFGHIJKL')
```

```python
# call-back function
greet()
```

```python
def greetPerson(name):
    print(f'Hello {name}')
```

```python
greetPerson('Charles')
```

```python
def greetReturn():
    return f'Hello World!!fghjklkjhg'
```

```python
greetReturn()
```

```python
def greetPersonReturn(name):
    return f'Hello {name}'
```

```python
greetPersonReturn('Bob')
```

```python
print(greetReturn())
```

```python
print(greet())
```

```python
'''
When to use Return type function?
-> If we need to call our function from one file to another then use reutrn type function.
-> we cannot use print function to display output. Just need to call function.
-> Only one return statement can be defined in one single function.

When to use non return type function?
-> If we need to call our function in same file where we defined it then use non return type function.
-> we can use print function to display output.
-> Multiple print statement can be defined in one single function.
'''
```

```python
def productDetails(product_name, product_desc, product_price, product_quantity):
    print(f'Product Name: {product_name}')
    print(f'Product Description: {product_desc}')
    print(f'Product Price: {product_price}')
    print(f'Product Quantity: {product_quantity}')
    print(f'Total Price: {product_price * product_quantity}')
```

```python
productDetails('Mobile', 'Black Mobile', 50000, 4)
```

## Default Argument Function

```python
def productDetails(product_name='Unknown', product_desc='Unknown Description', 
                   product_price=0, product_quantity=0):
    print(f'Product Name: {product_name}')
    print(f'Product Description: {product_desc}')
    print(f'Product Price: {product_price}')
    print(f'Product Quantity: {product_quantity}')
    print(f'Total Price: {product_price * product_quantity}')
```

```python
productDetails(100000)
```

## Keyword Argument Function

```python
def productDetails(product_name='Unknown', product_desc='Unknown Description', 
                   product_price=0, product_quantity=0):
    print(f'Product Name: {product_name}')
    print(f'Product Description: {product_desc}')
    print(f'Product Price: {product_price}')
    print(f'Product Quantity: {product_quantity}')
    print(f'Total Price: {product_price * product_quantity}')
```

```python
productDetails(product_quantity=10, product_name='Mobile', 
               product_desc = 'Black Mobile', product_price=2000)
```

## All Argument Function

```python
# *args
def personalDetails(name, *args):
    for data in args:
        print(data)
    print(name)
```

```python
personalDetails('Ram', 30, 'Bachelors', 'Lalitpur', 50000)
```

## All Keyword Argument Function

```python
# **kwargs
def kwPersonalDetails(**kwargs):
    print(kwargs)
```

```python
kwPersonalDetails(name='Ram', age=30, qualification= 'Bachelors', address='Lalitpur', salary=50000)
```
