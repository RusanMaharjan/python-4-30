```python
class BalanceWithdrawlError(Exception):
    pass

class NegativeAmountError(Exception):
    pass
```

```python
def balanceWithdraw():
    balance = 50000
    amount = int(input('Enter your withdrawl amount: '))

    if amount > 0:
        if amount < balance:
            balance -= amount
            print(f'Rs. {amount} has been withdrawn from A/C no. 1234567890')
        else:
            raise BalanceWithdrawlError(f'Error: Withdrawl amount must be less than balance amount.')
    else:
        raise NegativeAmountError(f'Error: Amount value must be positive integer.')
```

```python
try:
    balanceWithdraw()
except BalanceWithdrawlError as e:
    print(e)
except ValueError:
    print('Error: Only integer value are accepted by input field.')
except NegativeAmountError as e:
    print(e)
```
