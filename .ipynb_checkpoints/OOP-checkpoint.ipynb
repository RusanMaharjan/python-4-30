## Constructor

```python
'''
__init__()

Types of variable
---------------------
1. Local Variable -> Variable defined inside constructor/function as an argument.
2. Instance Variable -> Variable used to instantiate / store value of local variable.
                    -> Instance variable is also used to pass data from one class to another / one
                        object to another.
3. Global Variable -> Variable defined outside the class or function.
'''
```

```python
SCHOOL_NAME = 'ABC School'
class Student:
    def __init__(self, student_name, student_address, student_grade, student_phone):
        self.student_name = student_name
        self.student_address = student_address
        self.student_class = 'Class 9'
        self.student_grade = student_grade
        self.student_phone = student_phone

    def student_details(self):
        print(f'School Name: {SCHOOL_NAME}')
        print(f'Name of Student: {self.student_name}')
        print(f'Address of Student: {self.student_address}')
        print(f'Class of Student: {self.student_class}')
        print(f'Grade of Student: {self.student_grade}')
        print(f'Student Phone Number: {self.student_phone}')
```

```python
s = Student('Laswell', 'Jhapa', 'B+', 1234567890)
```

```python
s.student_details()
```

# Inheritance

```python
'''
- Parent Class and Child Class
- Child class inherit / extends parent class

Types of Inheritance
------------------------
1. Simple Inheritance (Two classes)
2. Multiple Inheritance (3 or more than 3 classes)
    -> One Parent, Multiple Child
3. Multi-level Inheritance (3 or more than 3 classes)
'''
```

```python
'''
- Rules of Inheritance
1. Child Class inherit Parent class.
2. If child class inherit parent class then child class needs to inherit parent class details as well.
'''
```

```python
# customer
# Parent Class
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone):
        self.customer_id = customer_id
        self.customer_name = customer_name
        self.customer_address = customer_address
        self.customer_phone = customer_phone

    def customer_details(self):
        print('Customer Details')
        print('-'*55)
        print(f'Customer Id: {self.customer_id}')
        print(f'Customer Name: {self.customer_name}')
        print(f'Customer Address: {self.customer_address}')
        print(f'Customer Phone: {self.customer_phone}')
```

```python
# Order
# Child Class
class Order(Customer):
    def __init__(self, order_id, order_date, order_price, order_destination, 
                 customer_id, customer_name, customer_address, customer_phone):
        super().__init__(customer_id, customer_name, customer_address, customer_phone)
        self.order_id = order_id
        self.order_date = order_date
        self.order_price = order_price
        self.order_destination = order_destination

    def order_details(self):
        print('Order Details')
        print('-'*55)
        print(f'Order Id: {self.order_id}')
        print(f'Order Date: {self.order_date}')
        print(f'Order Price: {self.order_price}')
        print(f'Order Destination: {self.order_destination}')
```

```python
o = Order(1, '2026-06-05', 'Rs. 5000', 'Lalitpur', 'Cus200', 'Bob', 'Bhaktapur', 1234567890)
```

```python
o.customer_details()
o.order_details()
```

```python
class Product(Customer):
    def __init__(self, product_id, product_name, product_desc, product_quantity, product_price,
                customer_id, customer_name, customer_address, customer_phone):
        super().__init__(customer_id, customer_name, customer_address, customer_phone)
        self.product_id = product_id
        self.product_name = product_name
        self.product_desc = product_desc
        self.product_quantity = product_quantity
        self.product_price = product_price

    def product_details(self):
        print('Product Details')
        print('-'*55)
        print(f'Product Id: {self.product_id}')
        print(f'Product Name: {self.product_name}')
        print(f'Product Description: {self.product_desc}')
        print(f'Product Quantity: {self.product_quantity}')
        print(f'Product Price: {self.product_price}')
```

```python
p = Product(100, 'Bag', 'Good Quality Bag', 5, 3000, 'Cus200', 'Bob', 'Bhaktapur', 1234567890)
```

```python
p.product_details()
```

## Multi-level inheritance

```python
# customer
# Grand-Parent Class
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone):
        self.customer_id = customer_id
        self.customer_name = customer_name
        self.customer_address = customer_address
        self.customer_phone = customer_phone

    def customer_details(self):
        print('Customer Details')
        print('-'*55)
        print(f'Customer Id: {self.customer_id}')
        print(f'Customer Name: {self.customer_name}')
        print(f'Customer Address: {self.customer_address}')
        print(f'Customer Phone: {self.customer_phone}')
```

```python
# Parent Class
class Product(Customer):
    def __init__(self, product_id, product_name, product_desc, product_quantity, product_price,
                customer_id, customer_name, customer_address, customer_phone):
        super().__init__(customer_id, customer_name, customer_address, customer_phone)
        self.product_id = product_id
        self.product_name = product_name
        self.product_desc = product_desc
        self.product_quantity = product_quantity
        self.product_price = product_price

    def product_details(self):
        print('Product Details')
        print('-'*55)
        print(f'Product Id: {self.product_id}')
        print(f'Product Name: {self.product_name}')
        print(f'Product Description: {self.product_desc}')
        print(f'Product Quantity: {self.product_quantity}')
        print(f'Product Price: {self.product_price}')
```

```python
# Child Class
class Order(Product):
    def __init__(self, order_id, order_date, order_price, order_destination,
                 product_id, product_name, product_desc, product_quantity, product_price,
                 customer_id, customer_name, customer_address, customer_phone):
        super().__init__(product_id, product_name, product_desc, product_quantity, product_price,
                         customer_id, customer_name, customer_address, customer_phone)
        self.order_id = order_id
        self.order_date = order_date
        self.order_price = order_price
        self.order_destination = order_destination

    def order_details(self):
        print('Order Details')
        print('-'*55)
        print(f'Order Id: {self.order_id}')
        print(f'Order Date: {self.order_date}')
        print(f'Order Price: {self.order_price}')
        print(f'Order Destination: {self.order_destination}')

        print('Product Details')
        print('-'*55)
        print(f'Product Id: {self.product_id}')
        print(f'Product Name: {self.product_name}')
        print(f'Product Description: {self.product_desc}')
        print(f'Product Quantity: {self.product_quantity}')
        print(f'Product Price: {self.product_price}')

        print('Customer Details')
        print('-'*55)
        print(f'Customer Id: {self.customer_id}')
        print(f'Customer Name: {self.customer_name}')
        print(f'Customer Address: {self.customer_address}')
        print(f'Customer Phone: {self.customer_phone}')
```

```python
o = Order(100, '2026-06-07', 'Rs.12000', 'Bhaktapur',
          1, 'Fridge', 'Black LG Fridge', 1, 'Rs.6000',
          'Cus100', 'Bob Marsten', 'Lalitpur', '34567898')
```

```python
# o.customer_details()
o.order_details()
```

## Encapsulation

```python
'''
- Data Hiding

Access Modifiers
--------------------
- Public (self.varName)
- Private (self.__varName)
- Protected (self._varName)
'''
```

```python
# School & Student
class School:
    def __init__(self, school_address, school_contact, school_branch):
        self.__school_name = 'ABC School'
        self.__school_address = school_address
        self.__school_contact = school_contact
        self.__school_branch = school_branch

    def __school_details(self):
        print(f'School Name: {self.__school_name}')
        print(f'School Address: {self.__school_address}')
        print(f'School Contact: {self.__school_contact}')
        print(f'School Branch: {self.__school_branch}')

    def details(self):
        self.__school_details()
```

```python
class Student(School):
    def __init__(self, student_id, student_name, student_contact, student_class, student_section,
                        school_address, school_contact, school_branch):
        super().__init__(school_address, school_contact, school_branch)
        self.__student_id = student_id
        self.__student_name = student_name
        self.__student_contact = student_contact
        self.__student_class = student_class
        self.__student_section = student_section

    def student_details(self):
        print(f'Student Id: {self.__student_id}')
        print(f'Student Name: {self.__student_name}')
        print(f'Student Contact: {self.__student_contact}')
        print(f'Student Class: {self.__student_class}')
        print(f'Student Section: {self.__student_section}')
```

```python
s = Student(100, 'Ram', 1234567890, 12, 'C13',
            'Lalitpur', 123498765, 'Balkumari')
```

```python
s.student_details()
```

```python
s.details()
```

### Protected

```python
class School:
    def __init__(self, school_address, school_contact, school_branch):
        self._school_name = 'ABC School'
        self._school_address = school_address
        self._school_contact = school_contact
        self._school_branch = school_branch

    def school_details(self):
        print(f'School Name: {self._school_name}')
        print(f'School Address: {self._school_address}')
        print(f'School Contact: {self._school_contact}')
        print(f'School Branch: {self._school_branch}')
```

```python
class Student(School):
    def __init__(self, student_id, student_name, student_contact, student_class, student_section,
                        school_address, school_contact, school_branch):
        super().__init__(school_address, school_contact, school_branch)
        self._student_id = student_id
        self._student_name = student_name
        self._student_contact = student_contact
        self._student_class = student_class
        self._student_section = student_section

    def student_details(self):
        print(f'Student Id: {self._student_id}')
        print(f'Student Name: {self._student_name}')
        print(f'Student Contact: {self._student_contact}')
        print(f'Student Class: {self._student_class}')
        print(f'Student Section: {self._student_section}')
```

## Polymorphism

```python
from datetime import datetime
class Messenger:
    def send_message(self, send_from, send_to, message):
        print('Messenger')
        print(f'From: {send_from}')
        print(f'To: {send_to}')
        print(f'Message: {message}')
        print(f'Sent: {datetime.now()}')

class Viber:
    def send_message(self, send_from, send_to, message):
        print('Viber')
        print(f'From: {send_from}')
        print(f'To: {send_to}')
        print(f'Message: {message}')
        print(f'Sent: {datetime.now()}')

class WhatsApp:
    def send_message(self, send_from, send_to, message):
        print('WhatsApp')
        print(f'From: {send_from}')
        print(f'To: {send_to}')
        print(f'Message: {message}')
        print(f'Sent: {datetime.now()}')

class Slack:
    def send_message(self, send_from, send_to, message):
        print('Slack')
        print(f'From: {send_from}')
        print(f'To: {send_to}')
        print(f'Message: {message}')
        print(f'Sent: {datetime.now()}')
```

```python
def message(obj, send_from, send_to, message):
    obj.send_message(send_from, send_to, message)
```

```python
message(Slack(), 'Ram', 'Shyam', 'Hello!!!')
```

## Abstraction

```python
'''
Abstraction -> Interface
- Hide implementation of code and details.
- If any class is made abstract class, atleast one function of it must be abstract function.
- Abstract Function -> pass keyword
- Other normal class implements abstract class.
- If any abstract class has only functions then it is also called as interface class.
- Interface class -> A class without body.
'''
```

```python
# Abstract Base Class
from abc import ABC, abstractmethod
```

```python
class Bank(ABC):
    @abstractmethod
    def userDetails(self, amount, account_number):
        pass

    @abstractmethod
    def connection(self):
        pass
```

```python
# Normal class implements abstract class
class MobileBanking(Bank):
    def mobileBankingConnection(self):
        return f'Connected to mobile banking server..'

    def userDetails(self, amount, account_number):
        return f'Rs.{amount} remaining in A/C no. {account_number}.'

    def connection(self):
        return f'Connected to Banking Database Server...'
```

```python
mb = MobileBanking()
```

```python
mb.userDetails(30000, 123456789056789)
```

```python
mb.mobileBankingConnection()
```

```python
mb.connection()
```
