#### Write a program that checks odd and even of 20 natural numbers.

```python
for i in range(1, 21):
    if i % 2 == 0:
        print(f'{i}: Even')
    else:
        print(f'{i}: Odd')
```

#### Match Case

```python
'''
1. You have ordered Breakfast.
2. You have ordered Lunch.
3. You have ordered Snaks.
4. You have ordered Dinner.
5. You have ordered Dessert.
6. Thank you for choosing us.
'''
```

```python
choose = int(input('Enter a number: '))

if choose == 1:
    print('You have ordered Breakfast.')
elif choose == 2:
    print('You have ordered Lunch.')
elif choose == 3:
    print('You have ordered Snaks.')
elif choose == 4:
    print('You have ordered Dinner.')
elif choose == 5:
    print('You have ordered Dessert.')
elif choose == 6:
    print('Thank you for choosing us.')
else:
    print('Invalid Number. Please choose order number from 1 to 5.\nClick 6 to exit program.')
```

```python
for i in range(5):
    choose = int(input('Enter a number: '))
    
    match choose:
        case 1:
            print('You have ordered Breakfast.')
        case 2:
            print('You have ordered Lunch.')
        case 3:
            print('You have ordered Snaks.')
        case 4:
            print('You have ordered Dinner.')
        case 5:
            print('You have ordered Dessert.')
        case 6:
            print('Thank you for choosing us.')
        case _:
            print('Invalid Number. Please choose order number from 1 to 5.\nClick 6 to exit program.')
else:
    print('Order limit!!')
```

```python
'''
Login Authentication System
-------------------------------
- Create a login system in which user can enter their credientials 3 times if each time their credientials are incorrect.
- If their input credintials are correct then program should display Login Successfull then terminate program.
- In case all 3 tries are failed then program should display "Account Blocked. Please contact admin support for password reset."
'''
```

```python
uname = 'abc@gmail.com'
pwd = '12345678'

for i in range(3):
    username = input('Enter your username: ')
    password = input('Enter your password: ')
    
    if username == uname and password == pwd:
        print('Login Successfull!!')
        break
    else:
        print('Try Again!!')
else:
    print('Account Blocked. Please contact admin support for password reset.')
```

## Break and Continue

```python
'''
- Break and Continue only can be used inside for or while loop.
- Break -> To stop program forcefully.
- Continue -> To skip values.
'''
# Print 10 natural numbers but stop program with it finds number 5.
for i in range(1, 11):
    if i == 5: # 5 == 5 -> True
        continue
    print(i)
```

## While Loop

```python
# Until condition becomes false.
secret_number = 7
guess_number = 0 # 7
remaining_attempt = 3

while guess_number != secret_number: # 7 != 7 -> False
    print(f'Remaining Attempt: {remaining_attempt}')
    print('-'*25)
    guess_number = int(input('Enter guess number: ')) # 7
    if guess_number == secret_number: # 7 == 7 # True
        print(f'Number {guess_number} guessed correctly.')
    else:
        print(f'Incorrect guessed number. Try again.\n')
        remaining_attempt -= 1 # remaining_attempt = remaining_attempt - 1

    if remaining_attempt == 0:
        print(f'{remaining_attempt} attempt left. Time Up!!')
        break
```

```python
# Write a program that: 
# Allows maximum 3 attempts 
# Continues asking for PIN using a while loop 
# Displays “Access Granted” if PIN is correct. 
# Displays “Card Blocked” after 3 incorrect attempts 
```

```python
pin = 1234
user_pin = 0000
balance = 50000
remaining_attempt = 3

while user_pin != pin:
    print(f'Remaining Attempt: {remaining_attempt}')
    user_pin = int(input('Enter your pin number: '))

    if user_pin == pin:
        while user_balance > balance:
            user_balance = int(input('Enter your balance amount: '))
            if user_balance < balance:
                if user_balance % 1000 == 0:
                    balance -= user_balance
                    print(f'{user_balance} has been withdrawn from A/C no. 1234567890')
                else:
                    print('Balance must be multiple of 1000s.')
            else:
                print(f'Withdraw balance must be less than current balance.')
                break
    else:
        print('Incorrect pin number!\n')
        remaining_attempt -= 1

    if remaining_attempt == 0:
        print('Pin Blocked. Contact Nearest Branch!!')
        break
```

```python
pin = 1234
user_pin = 0000
balance = 50000
remaining_attempt = 3

while remaining_attempt > 0:
    print(f'Remaining Attempt: {remaining_attempt}')
    user_pin = int(input('Enter your pin number: '))

    if user_pin == pin:
        user_balance = int(input('Enter your balance amount: '))
        if user_balance < balance:
            if user_balance % 1000 == 0:
                balance -= user_balance
                print(f'Rs.{user_balance} has been withdrawn from A/C no. 1234567890')
                break
            else:
                print('Balance must be multiple of 1000s.')
        else:
            print(f'Withdraw balance must be less than current balance.')
    else:
        print('Incorrect pin number!\n')
        remaining_attempt -= 1

    if remaining_attempt == 0:
        print('Pin Blocked. Contact Nearest Branch!!')
        break
```
