```python
'''
Variable Naming Convention
---------------------------
1. Snake Case
-> first_name, last_name, years_of_experience, monthly_total_funded_loan_amount_by_user 

2. Camel Case
-> firstName, lastName, yearOfExperience, monthlyTotalFundedLoanAmountByUser
'''
```

```python
'''
Operators
------------
1. Arithmetic Operator
-> (+, -, *, /, % (modulus / reminder), // (float division))

2. Logical Operator
-> AND(&), OR(|), NOT(!)

3. Comparision Operator
-> =, ==, >, >=, <, <=, !=
'''
```

```python
num1 = int(input('Enter first number: '))
num2 = int(input('Enter second number: '))

add = num1 + num2
sub = num1 - num2
mul = num1 * num2
div = num1 / num2
mod = num1 % num2
fd = num1 // num2

print(f'\nAddition of {num1} and {num2}: {add}')
print(f'Subtraction of {num1} and {num2}: {sub}')
print(f'Multiplication of {num1} and {num2}: {mul}')
print(f'Division of num1 and {num1} and {num2}: {div}')
print(f'Reminder/Modulus of {num1} and {num2}: {mod}')
print(f'Float Division of {num1} and {num2}: {fd}')
```

```python
# Logical Operator and Comparision Operator gives output in Boolean values (0/1).
```

```python
# =, ==, >, >=, <, <=, !=
# is num1 equals to num2?
num1 > num2
```
