```python
product_details_list
```

```python
## Loop to print dictionary values
'''
.keys() -> only to print keys of dictionary
.values() -> only to print values of dictionary
.items() -> To print both keys and values of dictionary 
'''
```

```python
for key in product_details_list.keys():
    print(key)
```

```python
for val in product_details_list.values():
    print(val)
```

```python
for key, val in product_details_list.items():
    print(f'{key}: {val}')
```

```python
for key, val in product_details_list.items():
    print(f'\n{key}')
    print('-'*15)
    for data in val:
        print(data)
```

```python
# Lab Question 5: The "Price Is Right" (Conditional Logic) You are given a list of product 
# prices: prices = [100, 250, 75, 400, 120, 310]. 
# 1. Write a program that creates two new empty lists: cheap_items and 
# expensive_items. 
# 2. Loop through the prices list. If a price is greater than 200, append it to 
# expensive_items. Otherwise, append it to cheap_items. 
# 3. Print both lists and calculate the total count of items in each using the len() 
# function.
```

```python
prices = [100, 250, 75, 400, 120, 310]
```

```python
cheap_items = []
expensive_items = []
```

```python
for price in prices:
    if price >= 200:
        expensive_items.append(price)
    else:
        cheap_items.append(price)
```

```python
print(cheap_items)
print(expensive_items)
```

```python
print(len(cheap_items))
print(len(expensive_items))
```
